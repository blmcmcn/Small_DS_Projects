```python
import cv2
import dlib
from imutils import face_utils
import os
from skimage import io
from PIL import Image, ImageOps
import numpy as np
from tqdm import tqdm
```

```python

cnn_face_detector = dlib.cnn_face_detection_model_v1("mmod_human_face_detector.dat")
```

```python
#finding faces and cropping them

save_directory = '/home/hadoop/DataChad/notebooks/cropped_images/'
get_directory = os.chdir('/home/hadoop/DataChad/notebooks/images')

for filename in tqdm(os.listdir(get_directory)):
        
    image = cv2.imread(filename)
    im = Image.open(filename)
    
    detected_faces = cnn_face_detector(image, 1)
    
    
    for n, face_rect in enumerate(detected_faces):
        #increase box size%
        width = face_rect.rect.right() - face_rect.rect.left()
        height = face_rect.rect.bottom() - face_rect.rect.top()
        
        x1 = face_rect.rect.left() - int(width*0.1)
        if x1 < 0:
            x1 = 0
        
        x2 = face_rect.rect.right() + int(width*0.13)
        if  x2 > image.shape[1]:
            x2 = int(image.shape[1]) - 1
            
        y1 = face_rect.rect.top() - int(height*0.16)
        if y1 < 0:
            y1 = 0
            
        y2 = face_rect.rect.bottom() + int(height*0.19)
        if  y2 > image.shape[0]:
            y2 = int(image.shape[0]) - 1
        
        box = (x1,y1,x2,y2)
        
        croppedFace = im.crop(box)
        
        new_filename = filename.replace(".jpg" , "_cropped_" + str(n) + ".jpg")
    
        croppedFace.save(save_directory + new_filename)
        
```

```python
#find the biggest cropped face for all images to deal with multiple faces

image_directory = '/home/hadoop/DataChad/notebooks/images/'
cropped_directory = '/home/hadoop/DataChad/notebooks/cropped_images/'
cropped_biggest_directory = '/home/hadoop/DataChad/notebooks/cropped_biggest_images/'

image_list = os.listdir(image_directory) 
number_files = len(image_list)

for i in tqdm(range(0, number_files)):
    
    string = str(i) + "_cropped_"
    
    faces_in_image = [filename for filename in os.listdir(cropped_directory) if filename.startswith(string)]
    
    size = []
    face_name = []
    
    if len(faces_in_image) > 0:
        
        for face in faces_in_image:
            width, height = Image.open(cropped_directory + face).size
            size.append(width*height)
            face_name.append(face)
        
        idx_largest_face = np.argmax(size)
    
        im = Image.open(cropped_directory + face_name[idx_largest_face])
        im_name = cropped_directory + face_name[idx_largest_face]
        
        new_filename = im_name.replace("cropped" , "cropped_biggest")
    
        im.save(new_filename)
```

```python
#finding the max width and height 

cropped_biggest_directory = '/home/hadoop/DataChad/notebooks/cropped_biggest_images/'

widths = []
heights = []

for filename in tqdm(os.listdir(cropped_biggest_directory)):

    im = Image.open(filename)
    
    width, height = im.size
    
    widths.append(width)
    heights.append(height)
    print(str(filename) + "    " + str(im.size))
    
maxXind = np.argmax(widths)
maxYind = np.argmax(heights)

print("\nmax width indices: \n")
print(maxXind)
print("\nmax height indices: \n")
print(maxYind)

max_width = widths[maxXind]
max_height = heights[maxYind]
print("\nmax widths: \n")
print(max_width)
print("\nmax heights: \n")
print(max_height)

```

```python
#resize the cropped images to the same size while maintaining aspect ratio 
#(requires padding)

max_size = max(max_width,max_height)

save_directory = '/home/hadoop/DataChad/notebooks/cropped_biggest_resized_images/'
get_directory = '/home/hadoop/DataChad/notebooks/cropped_biggest_images/'

for filename in tqdm(os.listdir(get_directory)):
    desired_size = max_size
    im = Image.open(get_directory + filename)
    old_size = im.size  # old_size[0] is in (width, height) format

    ratio = float(desired_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])
    im = im.resize(new_size, Image.ANTIALIAS)
    new_im = Image.new("RGB", (desired_size, desired_size))
    new_im.paste(im, ((desired_size-new_size[0])//2,
                    (desired_size-new_size[1])//2))


    new_filename = filename.replace("biggest" , "biggest_resized")
    new_im.save(save_directory + new_filename)
```
